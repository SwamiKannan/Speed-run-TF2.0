```python
import time
start_time=time.time()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_breast_cancer
```

```python
import pandas as pd
```

```python
base_data=load_breast_cancer()
```

```python
base_data.keys()
```

```python
features=base_data['data']
labels=base_data['target']
feature_names=base_data['feature_names']
target_names=base_data['target_names']
```

```python
df_feat=pd.DataFrame(features, columns=feature_names)
```

```python
print(df_feat.isna().sum())
```

```python
df_feat.head()
```

```python
df_feat.describe()
```

```python
X_train, X_test, y_train, y_test=train_test_split(features,labels,shuffle=True, stratify=labels,test_size=0.33)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Input, Dense
```

```python
model=Sequential([ Input(shape=X_train.shape[1],),
                 Dense(1, activation='sigmoid')])
```

```python
model.compile(optimizer='Adam', loss='binary_crossentropy',metrics=['accuracy'])
```

```python
mid_time=time.time()
```

```python
history=model.fit(X_train,y_train, epochs=100,validation_data=(X_test,y_test))
```

```python
end_time=time.time()
```

```python
model.evaluate(X_train,y_train)
model.evaluate(X_test,y_test)
```

```python
history.history.keys()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(100),history.history['loss'],'r')
plt.plot(range(100),history.history['val_loss'],'b')
plt.show()
```

```python
plt.plot(range(100),history.history['accuracy'],'r')
plt.plot(range(100),history.history['val_accuracy'],'b')
plt.show()
```

```python
final_end_time=time.time()
```

```python
print('Time for: \n')
print('EDA + preprocessing + modelling', (mid_time-start_time)/60,'minutes')
print('Till model training', (end_time-start_time)/60,'minutes')
print('Till model evaluation', (final_end_time-start_time)/60,'minutes')
```
